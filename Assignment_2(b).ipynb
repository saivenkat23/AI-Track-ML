```
#intializing parameters
x_o=2
eta=0.03
eps=0.000001
del_x=1
max_iters=1000
iters=0
y,z=0,0

def deriv(x):
  y=x**3
  x_deriv = 4*y+6*x 
  return x_deriv

while abs(del_x)>eps and iters<max_iters:
  prev_x=x_o
  del_x=-eta*deriv(prev_x)
  x_o=x_o+del_x
  iters=iters+1
  print("Iteration",iters,"\nX value is ",x_o)

print("The Local minimum occurs at ",x_o)

```
